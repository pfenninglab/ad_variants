```python
import pandas as pd
import numpy as np
```

```python
def oneHotEncodeSequence(sequence):
    oneHotDimension = (len(sequence), 4)
    dnaAlphabet = {"A":0, "G":1, "C":2, "T":3}
    one_hot_encoded_sequence = np.zeros(oneHotDimension, dtype=np.int)
    for i, nucleotide in enumerate(sequence):
        if nucleotide.upper() in dnaAlphabet:
            index = dnaAlphabet[nucleotide.upper()]
            one_hot_encoded_sequence[i][index] = 1
    return one_hot_encoded_sequence
```

```python
snp_data = pd.read_csv("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/snigdha_snp_list_unique_haploreg_hg19_positions_1kb_and_500bp_sequences.txt",
                       header=0,
                       sep='\t')
```

```python
snp_data.head()
```

```python
encodedRefSequences_1kb = []
encodedAltSequences_1kb = []
encodedRefSequences_500bp = []
encodedAltSequences_500bp = []

ref_1kb_output = "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/ref_1kb_sequences.npy"    
alt_1kb_output = "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/alt_1kb_sequences.npy"    
ref_500bp_output = "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/ref_500bp_sequences.npy"  
alt_500bp_output = "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/alt_500bp_sequences.npy"  


for index,row in snp_data.iterrows():
    refSequence_1kb = row["ref_1kb"]
    altSequence_1kb = row["alt_1kb"]
    refSequence_500bp = row["ref_500bp"]
    altSequence_500bp = row["alt_500bp"]
    
    encodedRefSequences_1kb.append(oneHotEncodeSequence(refSequence_1kb))
    encodedAltSequences_1kb.append(oneHotEncodeSequence(altSequence_1kb))
    encodedRefSequences_500bp.append(oneHotEncodeSequence(refSequence_500bp))
    encodedAltSequences_500bp.append(oneHotEncodeSequence(altSequence_500bp))
    
encodedRefSequences_1kb = np.stack(encodedRefSequences_1kb, axis=0)
encodedAltSequences_1kb = np.stack(encodedAltSequences_1kb, axis=0)
encodedRefSequences_500bp = np.stack(encodedRefSequences_500bp, axis=0)
encodedAltSequences_500bp = np.stack(encodedAltSequences_500bp, axis=0)


np.save(ref_1kb_output, encodedRefSequences_1kb)
np.save(alt_1kb_output, encodedAltSequences_1kb)
np.save(ref_500bp_output, encodedRefSequences_500bp)
np.save(alt_500bp_output, encodedAltSequences_500bp)
```
