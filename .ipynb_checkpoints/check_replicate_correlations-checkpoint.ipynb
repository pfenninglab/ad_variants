```python
import pandas as pd
import argparse
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
rcParams['svg.fonttype'] = 'none'
rcParams['font.size']=15
```

```python
inFiles = {"GM12878": "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/lcl_dnase_encode/atac/5840747c-53d2-4b55-8091-72684bd9fc7f/call-idr/shard-0/execution/rep1_rep2.idr0.05.thresholded-peaks.bfilt.txt.gz",
           "Monocyte": "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_human_monocyte_brain/furtherProcessedPeaks/humanMonocyteDHS/peak/macs2/idr/pooled_pseudo_reps/humanMonocyteDHS_ppr.IDR0.1.thresholded-peaks.bfilt.txt.gz",
           "Microglia": "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/gosselin_microglia_atac/exvivo_idr_pipeline_out/atac/0c4b0010-a158-4176-a424-059a4fb12cef/call-idr_ppr/execution/ppr.idr0.05.thresholded-peaks.bfilt.txt.gz",
           "Neuron": "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_fullard_neun/ppr.idr0.1.thresholded-peaks.bfilt.txt.gz"
          }

repNotPpr = {"GM12878": True,
             "Monocyte": False,
             "Microglia": False,
             "Neuron": False
            }
```

```python
def plotData(inFiles, repNotPpr, cellType):
    inFile = inFiles[cellType]
    replicate_data = pd.read_csv(inFile,
                                 sep="\t", compression="gzip",
                                 names=["CHR",
                                 "START",
                                 "END",
                                 "NAME",
                                 "SCORE",
                                 "STRAND",
                                 "SIGNAL",
                                 "P",
                                 "Q",
                                 "SUMMIT",
                                 "LOCALIDR",
                                 "GLOBALIDR",
                                 "REP1_START",
                                 "REP1_END",
                                 "REP1_SIGNAL",
                                 "REP1_SUMMIT",
                                 "REP2_START",
                                 "REP2_END",
                                 "REP2_SIGNAL",
                                 "REP2_SUMMIT"]
                                )
    print("Overall correlations")
    print(replicate_data.shape)
    print(pearsonr(replicate_data["REP1_SIGNAL"], replicate_data["REP2_SIGNAL"]))
    print(spearmanr(replicate_data["REP1_SIGNAL"], replicate_data["REP2_SIGNAL"]))
    
    
    chrs_training_replicate_data = replicate_data.loc[~replicate_data["CHR"].isin(["chr4", "chr8", "chr9"])]
    print("Training chr correlations")
    print(chrs_training_replicate_data.shape)
    print(pearsonr(chrs_training_replicate_data["REP1_SIGNAL"], chrs_training_replicate_data["REP2_SIGNAL"]))
    print(spearmanr(chrs_training_replicate_data["REP1_SIGNAL"], chrs_training_replicate_data["REP2_SIGNAL"]))
    
    
    chr4_replicate_data = replicate_data.loc[replicate_data["CHR"]=="chr4"]
    print("chr4 correlations")
    print(chr4_replicate_data.shape)
    print(pearsonr(chr4_replicate_data["REP1_SIGNAL"], chr4_replicate_data["REP2_SIGNAL"]))
    print(spearmanr(chr4_replicate_data["REP1_SIGNAL"], chr4_replicate_data["REP2_SIGNAL"]))
    
    
    addStr = "ppr"
    if repNotPpr[cellType]:
        addStr = "rep"
    
    plt.scatter(chr4_replicate_data["REP1_SIGNAL"],
                chr4_replicate_data["REP2_SIGNAL"],
                s=5,
                alpha=0.5,
                rasterized=True
               )
    plt.ylabel("Open chromatin signal ("+cellType+" "+addStr+"1)")
    plt.xlabel("Open chromatin signal ("+cellType+" "+addStr+"2)")
    plt.show()
    #plt.savefig(cellType + "_replicate_concordance.svg")
    #plt.close()
```

```python
for cellType in inFiles:
    plotData(inFiles, repNotPpr, cellType)
```
