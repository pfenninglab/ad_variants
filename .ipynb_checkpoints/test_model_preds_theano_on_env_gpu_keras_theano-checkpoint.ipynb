```python
#Before using this script, make sure backend is set to "theano" in $HOME/.keras/keras.json and make sure that gpu_keras_theano environment is activated

from keras.models import load_model
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
monocyte_regression = load_model("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_gosselin_microglia/model_7.hdf5", compile = False)
```

```python
test_sequences = np.load("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_gosselin_microglia/chr4_summit_centered_validation_set_X.npy")
y_true = np.load("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ml_gosselin_microglia/chr4_summit_centered_validation_set_Y.npy")
```

```python
pred = monocyte_regression.predict(test_sequences)
```

```python
pearsonr(y_true, pred)
```

```python
plt.scatter(y_true, pred, s=5)
```
