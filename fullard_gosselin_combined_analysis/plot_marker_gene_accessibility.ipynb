```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
rcParams['svg.fonttype'] = 'none'
rcParams['font.size']=10
```

```python
gosselin_peaks_homer = pd.read_csv("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/gosselin_fullard_combined_analysis/gosselin_microglia_exvivo_idr_optimal_peak_homer_annot.txt", sep="\t")
fullard_peaks_homer = pd.read_csv("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/gosselin_fullard_combined_analysis/fullard_PUT_neunplus_idr_optimal_peak_homer_annot.txt", sep="\t")
```

```python
gosselin_peaks_narrowPeak = pd.read_csv("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/gosselin_fullard_combined_analysis/gosselin_microglia_exvivo_idr_optimal_peak.narrowPeak", sep="\t", header=None)
fullard_peaks_narrowPeak = pd.read_csv("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/gosselin_fullard_combined_analysis/fullard_PUT_neunplus_idr_optimal_peak.narrowPeak", sep="\t", header=None)
```

```python
gosselin_peaks_homer["SIGNAL"] = np.log10(gosselin_peaks_narrowPeak[6])
fullard_peaks_homer["SIGNAL"] = np.log10(fullard_peaks_narrowPeak[6])
```

```python
habib_expfc1_markers = pd.read_csv("/projects/pfenninggroup/jemmie/gwyn_hpc_3exphg/new_habib_marker_sets/exPFC1_markers.txt", names=["exPFC1"])
habib_expfc2_markers = pd.read_csv("/projects/pfenninggroup/jemmie/gwyn_hpc_3exphg/new_habib_marker_sets/exPFC2_markers.txt", names=["exPFC2"])
habib_mg_markers = pd.read_csv("/projects/pfenninggroup/jemmie/gwyn_hpc_3exphg/new_habib_marker_sets/MG_markers.txt", names=["MG"])
```

```python
gosselin_expfc1 = gosselin_peaks_homer.loc[gosselin_peaks_homer["Gene Name"].isin(habib_expfc1_markers["exPFC1"])]
fullard_expfc1 = fullard_peaks_homer.loc[fullard_peaks_homer["Gene Name"].isin(habib_expfc1_markers["exPFC1"])]
```

```python
gosselin_expfc2 = gosselin_peaks_homer.loc[gosselin_peaks_homer["Gene Name"].isin(habib_expfc2_markers["exPFC2"])]
fullard_expfc2 = fullard_peaks_homer.loc[fullard_peaks_homer["Gene Name"].isin(habib_expfc2_markers["exPFC2"])]
```

```python
gosselin_mg = gosselin_peaks_homer.loc[gosselin_peaks_homer["Gene Name"].isin(habib_mg_markers["MG"])]
fullard_mg = fullard_peaks_homer.loc[fullard_peaks_homer["Gene Name"].isin(habib_mg_markers["MG"])]
```

```python
plt.violinplot([fullard_mg["SIGNAL"].tolist(), gosselin_mg["SIGNAL"].tolist()])
plt.boxplot([fullard_mg["SIGNAL"].tolist(), gosselin_mg["SIGNAL"].tolist()])
plt.show()
```

```python
plt.violinplot([fullard_expfc1["SIGNAL"].tolist(), gosselin_expfc1["SIGNAL"].tolist()])
plt.boxplot([fullard_expfc1["SIGNAL"].tolist(), gosselin_expfc1["SIGNAL"].tolist()])
plt.show()
```

```python
plt.violinplot([fullard_expfc2["SIGNAL"].tolist(), gosselin_expfc2["SIGNAL"].tolist()])
plt.boxplot([fullard_expfc2["SIGNAL"].tolist(), gosselin_expfc2["SIGNAL"].tolist()])
plt.show()
```

```python
plt.violinplot([fullard_peaks_homer["SIGNAL"].tolist(), gosselin_peaks_homer["SIGNAL"].tolist()])
plt.boxplot([fullard_peaks_homer["SIGNAL"].tolist(), gosselin_peaks_homer["SIGNAL"].tolist()])
plt.show()
```
