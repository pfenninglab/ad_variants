```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
rcParams['svg.fonttype'] = 'none'
rcParams['font.size']=10
```

```python
gosselin_qc_info_file = '/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/gosselin_microglia_atac/gosselin_atac_pipeline_out/atac/gosselin_qc.tsv'
gosselin_qc_info = pd.read_csv(gosselin_qc_info_file, sep="\t", skiprows=1)
```

```python
fig, ax = plt.subplots()

parts = ax.violinplot(dataset=[gosselin_qc_info.num_reads])

parts['bodies'][0].set_facecolor('r')
parts['bodies'][0].set_edgecolor('black')
parts['bodies'][0].set_alpha(1)

ax.scatter([1]*gosselin_qc_info.shape[0], gosselin_qc_info.num_reads)

plt.boxplot(x=[gosselin_qc_info.num_reads])
plt.show()
```

```python
fig, ax = plt.subplots()

parts = ax.violinplot(dataset=[gosselin_qc_info["N_opt.1"]])

parts['bodies'][0].set_facecolor('r')
parts['bodies'][0].set_edgecolor('black')
parts['bodies'][0].set_alpha(1)


ax.scatter([1]*gosselin_qc_info.shape[0], gosselin_qc_info["N_opt.1"])


plt.boxplot(x=[gosselin_qc_info["N_opt.1"]])
plt.show()
```

```python
print(np.mean(gosselin_qc_info["N_opt.1"]))
print(np.std(gosselin_qc_info["N_opt.1"]))
print(np.median(gosselin_qc_info["N_opt.1"]))
print(np.min(gosselin_qc_info["N_opt.1"]))
print(np.max(gosselin_qc_info["N_opt.1"]))
```
