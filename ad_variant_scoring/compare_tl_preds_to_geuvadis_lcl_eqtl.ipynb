```python
eqtl_catalogue_geuvadis_lcl = pd.read_csv("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/geuvadis_lcl_eqtl/QTD000110.all.tsv.gz",
                                          compression='gzip',
                                          sep='\t'
                                         )
```

```python
genome_wide_cutoff = 5e-4
eqtl_catalogue_geuvadis_lcl_sig = eqtl_catalogue_geuvadis_lcl.loc[eqtl_catalogue_geuvadis_lcl['pvalue']<genome_wide_cutoff]
```

```python
tanzi_2m_downsample_variant_info_gm_overlapping= pd.read_csv("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/geuvadis_lcl_eqtl/tanzi_2m_downsample_tl_model_scores_and_gm_overlaps.txt",
                                                       sep='\t',
                                                      )
```

```python
tanzi_variant_info_eqtl_merged = tanzi_2m_downsample_variant_info_gm_overlapping.merge(eqtl_catalogue_geuvadis_lcl_sig,
                                                      left_on='RSID',
                                                      right_on='rsid',
                                                      suffixes=['_orig', '_eqtlcat'],
                                                      how='inner'
                                                     )
```

```python
coef = np.polyfit(tanzi_variant_info_eqtl_merged['DNase_effect_score'],
                 tanzi_variant_info_eqtl_merged['TLavg_effect_score'],
                 1)

poly1d_fn = np.poly1d(coef) 

plt.plot(tanzi_variant_info_eqtl_merged['DNase_effect_score'],
         tanzi_variant_info_eqtl_merged['TLavg_effect_score'],
         'C0o',
         tanzi_variant_info_eqtl_merged['DNase_effect_score'],
         poly1d_fn(tanzi_variant_info_eqtl_merged['DNase_effect_score']),
         '--k')
plt.ylabel('TLavg_effect_score')
plt.xlabel('DNase_effect_score')
```

```python
def shortest_distance(x1,y1, m, b):
    numerator = m*x1+b-y1
    denominator = np.sqrt(m**2+1)
    return numerator/denominator
```

```python
tanzi_variant_info_eqtl_merged["shortest_dist"] = np.inf
for i,row in tanzi_variant_info_eqtl_merged[['DNase_effect_score',
                                           'TLavg_effect_score']].iterrows():
    tanzi_variant_info_eqtl_merged.at[i,'shortest_dist'] = shortest_distance(row['DNase_effect_score'],
                                                                             row['TLavg_effect_score'],
                                                                             coef[0],
                                                                             coef[1]
                                                                            )
```

```python
sns.scatterplot(x='beta', y='shortest_dist', data= tanzi_variant_info_eqtl_merged)
```
