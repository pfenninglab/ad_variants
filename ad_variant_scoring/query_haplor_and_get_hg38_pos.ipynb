```r
library(haploR)
```

```r
snp_data <- read.table("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/snigdha_snp_list_unique.txt",
                       sep='\t',
                       header=TRUE)
```

```r
rsids <- as.vector(snp_data[["LD_RSID"]])
```

```r
rsids
```

```r
extended_data <- getExtendedView(snp='rs34388456')
```

```r
hg38_positions = c()
```

```r
for (rsid in rsids) {
    extended_data <- getExtendedView(snp=rsid)
    hg38_pos <- as.integer(as.character(extended_data[[1]]$`Sequence facts`$`pos (hg38)`))
    hg38_positions <- c(hg38_positions, hg38_pos)
}
```

```r
hg38_positions
```

```r
snp_data["haploreg_hg38_pos"] <- hg38_positions
```

```r
snp_data_relevant_columns <- subset(snp_data, select = c("CHR","haploreg_hg38_pos","LD_RSID","REF","ALT") )
```

```r
write.table(snp_data_relevant_columns,
            file = "/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/snigdha_snp_list_unique_haploreg_hg38_positions.txt",
            sep = "\t",
            quote = FALSE,
            row.names=FALSE)
```
