```r
library(haploR)
```

```r
snp_data <- read.table("/projects/pfenninggroup/machineLearningForComputationalBiology/eramamur_stuff/ad_variants_processing/snigdha_snp_list_unique.txt",
                       sep='\t',
                       header=TRUE)
```

```r
rsids <- as.vector(snp_data[["LD_RSID"]])
```

```r
hg19_positions = c()
```

```r
for (rsid in rsids) {
    extended_data <- getExtendedView(snp=rsid)
    hg19_pos <- as.integer(as.character(extended_data[[1]]$`Sequence facts`$`pos hg19`))
    hg19_positions <- c(hg19_positions, hg19_pos)
    reference <- as.character(extended_data[[1]]$`Sequence facts`$`Reference`)
    alternate <- as.character(extended_data[[1]]$`Sequence facts`$`Alternate`)
    #cat(snp_data[row,"REF"])
    cat(rsid)
    cat(" ")
    cat(hg19_pos)
    cat(" ")
    cat(reference)
    cat(" ")
    #cat(snp_data[row,"ALT"])
    cat(alternate)
    print("")
}
```
